```python
## Database에서 sql로 데이터를 추출해서 pandas의 DataFrame으로 
## 생성
## 사용하는 DBMS는 MySQL, 사용하는 데이터는 book table
## DBMS는 강사용 컴퓨터를 이용 IP = 70.12.16.160
# conda install pymysql
import pymysql.cursors # MySQL DBMS 사용하기 위한 module 로딩
import numpy as np
import pandas as pd

# 데이터베이서 connection

conn = pymysql.connect(host="localhost",
                       user="data",
                       password="data",
                       db="library",
                       charset="utf8")

keyword = "여행"
# database에서 data를 가져오기 위한 SQL을 작성
# 책 제목의 keyword가 들어있는 행을 찾아서 제목 가격 저자정보를 가져온다!
sql = "select btitle,bprice,bauthor " + \
        "from book " + \
        "where btitle like '%{}%'".format(keyword)

try:
    df = pd.read_sql(sql, conn)
    display(df)
    # 정상적으로 DataFrame을 얻어왔으면 
    # JSON File로 저장
    df.to_json("./data/book_columns.json", orient="columns")
    df.to_json("./data/book_records.json", orient="records")
    df.to_json("./data/book_index.json", orient="index")
    df.to_json("./data/book_values.json", orient="values")
    
    
except Exception as inst:
    print(inst)
finally:
    conn.close()
    print("=="*20)
```

```python
## 영화진흥위원회의 일일 박스오피스 정보를 이용해서
## DataFrame을 만들어 보자!
## 접속해서 JSON을 얻어내면 그걸 분석해서
## 내가 원하는 형태의 DataFrame을 생성

import numpy as np
import pandas as pd
import urllib # network접속을 하기 위해서 필요
import json   # open API의 결과가 JSON을 읽어들이기 위한 module

k_val = "3722f3397a7e0428f41e02c489828eee"
n_date = "20191120"

movie_url = "http://www.kobis.or.kr/kobisopenapi/webservice/"+\
            "rest/boxoffice/searchDailyBoxOfficeList.json?"+\
            "key={}&targetDt={}".format(k_val, n_date)

m_page = urllib.request.urlopen(movie_url) # 결과 JSON 문자열이 들어있는 page 객체
print(m_page)

json_page = json.loads(m_page.read())


### 이렇게 JSON을 얻었으면 필요한 부분만 추려서 DataFrame으로
### 생성하면 된다!

## 순위, 영화제목, 당일매출액
## DataFrame으로 구성해서 작성해 보자.

n_data = json_page["boxOfficeResult"]['dailyBoxOfficeList']
df = pd.DataFrame(n_data)
rdf = pd.DataFrame(n_data, columns=["rank","movieNm","salesAmt"])
cut_df = df[["rank","movieNm","salesAmt"]]
#cut_df.columns.names("순위","영화제목","일 매출액")
display(cut_df)
display(rdf)
```

```python
%%time

movie_data = {}
m_dict = {}
rank_li = []
title_li = []
sales_li = []

for m_dict in json_page["boxOfficeResult"]['dailyBoxOfficeList']:
    rank_li.append(m_dict["rank"])
    title_li.append(m_dict["movieNm"])
    sales_li.append("{:,}".format(int(m_dict["salesAmt"])))
    
movie_data["순위"] = rank_li
movie_data["영화제목"] = title_li
movie_data["일 매출액"] = sales_li
ndf = pd.DataFrame(movie_data)
display(ndf)
```

```python
## 공분산 
## 공분산 : 두개의 확률변수 X, Y가 있을 때
##         이 두개의 확률변수의 관계를 보여줄 때 사용한다!
## 공분산은 두 확률변수의 편차의 곱의 평균으로 표현
## 데이터가 평균으로부터 얼마나 떨어져 있는지 표현!
## 그래프를 그려 확인해보자

# conda install matplotlib

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(4) # 항상 같은 난수가 발생하도록 seed 처리!

x = np.random.randint(-20, 20, (10,))
y = np.random.randint(-10, 10, (10,))
x_mean = x.mean()
y_mean = y.mean()
print(x,y)

## scatter(산점도)
plt.scatter(x,y,color="red")
plt.scatter(x_mean,y_mean,color="blue")
plt.show()

x_cov = x-x_mean
y_cov = y-y_mean
print(x_cov)
print(y_cov)
result = 0
for i in range(10):
    result += x_cov[i] * y_cov[i]
myCov = result/9 # 모공분산 공식
                  # 표준공분산 : (n-1)로 나누어 준다
print(myCov)
print(np.cov(x,y))
```

```python
## DataFrame에서 sum함수와 mean함수의 사용

import numpy as np
import pandas as pd

## 중첩리스트 이용해서 DataFrame을 생성해 보자

data = [[2, np.nan],
        [7, -3],
        [np.nan, np.nan],
        [1, -2]]
df = pd.DataFrame(data,
                 columns=["one", "two"],
                 index=["a", "b", "c", "d"])
display(df)

# df.sum(axis=0) # DataFrame에서는 sum()함수에 대한 axis를 주지 않으면
#                # axis=0으로 세팅. default = 0
#                # NaN을 0으로 간주해서 계산
# df.sum(axis=1)        

# df.mean()

# df.loc[["a","b"]].sum()

# df.mean(axis=0, skipna=True) # "skipna=False" NaN은 실수처리 -> 모든연산결과는 NaN

## "one" column의 결측값은 "one" column의 평균으로
## "two" column의 결측값은 "two" column의 최소값으로 대채해 보아요
df["one"] == np.nan

one_mean = df["one"].mean(axis=0, skipna=True)
two_min = df["two"].min(axis=0, skipna=True)

print( one_mean, two_min)

df["one"] = df["one"].fillna(value=one_mean)
df["two"] = df["two"].fillna(value=two_min)
display(df)
```

```python
## DataFrame Merge
## inner join, outer join(full), left join, right join

import numpy as np
import pandas as pd

data1 = {
    "학번" : [1,2,3,4],
    "이름" : ["홍길동","최길동","아이유","김연아"],
    "학년" : [2, 3, 3, 1]
}
data2 = {
    "학번" : [1,2,4,5],
    "학과" : ["컴퓨터","미술","철학","사회"],
    "학점" : [1.3, 3.5, 4.3, 2.3]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)

pd.merge(df1,df2, on="학번", how="outer") # inner, outer, left, right
```

```python
## inner join 예1)
import numpy as np
import pandas as pd

data1 = {
    "학번" : [1,2,3,4],
    "이름" : ["홍길동","최길동","아이유","김연아"],
    "학년" : [2, 3, 3, 1]
}
data2 = {
    "학생번호" : [1,2,4,5],
    "학과" : ["컴퓨터","미술","철학","사회"],
    "학점" : [1.3, 3.5, 4.3, 2.3]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)

pd.merge(df1,df2, left_on="학번", 
         right_on="학생번호",
         how="inner") # 출력결과가 left_ right_ 같이나옴!
```

```python
## inner join 예2)
import numpy as np
import pandas as pd

data1 = {
    "학번" : [1,2,3,4],
    "이름" : ["홍길동","최길동","아이유","김연아"],
    "학년" : [2, 3, 3, 1]
}
data2 = {
    
    "학과" : ["컴퓨터","미술","철학","사회"],
    "학점" : [1.3, 3.5, 4.3, 2.3]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df2.index = [1,2,4,5]


pd.merge(df1,df2, left_on="학번", 
         right_index=True,
         how="inner") # 출력결과가 left_on 과 right의 index와 mapping해서 출력
```

```python
## inner join 예3)
import numpy as np
import pandas as pd

data1 = {
    
    "이름" : ["홍길동","최길동","아이유","김연아"],
    "학년" : [2, 3, 3, 1]
}
data2 = {
    
    "학과" : ["컴퓨터","미술","철학","사회"],
    "학점" : [1.3, 3.5, 4.3, 2.3]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df1.index = [1,2,3,4]
df2.index = [1,2,4,5]


pd.merge(df1,df2, 
         left_index=True, 
         right_index=True,
         how="inner") # 출력결과가 left와 right의 index와 mapping해서 출력
```
