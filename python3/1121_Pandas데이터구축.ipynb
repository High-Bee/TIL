```python
# 파이썬 문법
# 10/0 , 보통 10을 0으로 나누면 무한대가 출력되지만 파이썬은수행되지 않고 중지된다
# 의도치 않은 에러가 발생되는 경우 
# 프로그래밍 단계에서는 문제가 없지만 실행시 문제가 발생할 수 있는 코드가 있다
# 네트워크 연결이 안되서 오픈 API 불러올 수 없을 때도 에러 발생
# 이런 상황들을 통칭해서 exception이라고 한다 (프로그램 error와는 다른 개념)
# 전원이 나가면, fatal error / 복구가 가능한 (파일이 없거나,,네트워크 끊기거나) 에러는 exception
# exception 상황의 경우 프로그램이 멈추지 않기때문에 exception 관리 가능하다

# exceptiong 처리 : exception handling
# 위에서 exception이 발생해도 아래 구문이 출력되도록 처리
try:
    10/0
except Exception as inst:
    print(inst) # 예외 상황에 대한 내용을 출력
finally:
    print("호호")

# finally 문제가 있던 없던 무조건 출력
# 프로그램은 계속 진행된다
```

```python
# OPEN API를 이용해서 JSON으로부터 dataframe 생성
# 영화 진흥회의 일일박스오피스 정보를이용해서 dataframe생성
# 접속해서 JSON을 얻어내면 분석해서 원하는 형태의 dataframe 생성

import numpy as np
import pandas as pd
# network 접속할때 필요한 모듈
import urllib
# open api결과가 json일때 dict로 변경하는 모듈
import json

movie_url = "http://www.kobis.or.kr/kobisopenapi/webservice/" + \
            "rest/boxoffice/searchDailyBoxOfficeList.json?key={}&targetDt={}".format("492b365825ea24ec65fb822d02968a09",
                                            "20191120")

page = urllib.request.urlopen(movie_url) #주소가 있는 객체를 만들고

json_page = json.loads(page.read()) #JSON 문자열 읽어서 dict로 변환
print(json_page)

# 이렇게 데이터를 얻었으면 필요한 부분만 추려서 dataframe 생성
# 순위, 영화제목, 판매량, 금액, 당일 매출액을 데이터프레임으로 구성
# 작성해보기

display(json_page)

# 키값 : rank, movieNm, salesamt, 

movie_data = {}
rank_list = list()
title_list = list()
sales_list = list()

for m_dict in json_page["boxOfficeResult"]["dailyBoxOfficeList"]:
    rank_list.append(m_dict["rank"])
    title_list.append(m_dict["movieNm"])
    sales_list.append("{:}".format(int(m_dict["salesAmt"])))
                      
movie_data["순위"] = rank_list
movie_data["영화제목"] = title_list
movie_data["당일매출액"] = sales_list
                      
df = pd.DataFrame(movie_data)
display(df)
```
