```python
## python에 남은 built-in Type

## Set Type ( set )
## 어떻게 생겻나요? -> {}
## {} : 기본적으로 dict, dict는 {}안에 key와 value로 구성
## set은 {} 안에 요소만 들어간다!
## set은 순서가 없고 중복을 허용하지 않는 저장장소

# 간단한 예제

a = set([1,2,3,4,5,1,2,3,1,23,1,41])

a = set("Hello") # 문자열을 set으로 만들면 {'e', 'o', 'l', 'H'}

print(a)

# set 연산

a = {1,2,3,4} # set의 literal
b = {3,4,5,6} 




a.intersection(b)
print(a & b) # 교집합 연산 : set에 & 적용하면 교집합
a.union(b)
print(a | b) # 합집합 연산 : set에 | 적용하면 합집합
a.difference(b)
print(a - b) # 차집합 연산 : set에 - 적용하면 차집합

mySet = {1,2,3,4,5}
mySet.add(7)  # add()는 set에 원하는 요소 1개를 추가

print(mySet)

mySet.update([10,11,12]) # 여러가지를 추가
print(mySet)

mySet.remove(1) # 삭제
print(mySet)
```
